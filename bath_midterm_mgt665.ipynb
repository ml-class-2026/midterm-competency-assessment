```python
## Midterm Competency Assessment - Spring 2026
## Problem 1. Run a linaer regression (OLS)
# This model uses credit score and debt to income (DTI) ratio to try to predict whether or not a 
# loan is approved. In the dataset 1 means approved and 0 means denied.   
```

```python
# This section loads the loan dataset and prints out the headers with a few examples, 
# as well as the dataset's statistics.
```

```python
import pandas as pd
import statsmodels.api as sm

df = pd.read_csv('loan_approval.csv')
print(df.head())
print(df.describe())
```

```python
# This section sets our predictor variables (X - DTI and credit score), and our outcome variable (Y - approved
# loan or not). A constant is added to calculate the intercept. 
```

```python
x = df[['credit_score', 'debt_to_income']]
x = sm.add_constant(x)

y = df['approved']
```

```python
# This section uses OLS regression (OLS) using credit score and DTI ratio to predict whether a loan is
# approved or not. It then prints the model summary. 
```

```python
ols_model = sm.OLS(y, x).fit()
print(ols_model.summary())
```

```python
## Problem 2. Run a logistic regression
# This section of the code runs a logistic regression model (Logit) with the same predictor variables in order
# to predict loan approval. It then prints the model summary for analysis. 
```

```python
logit_model = sm.Logit(y, x).fit()
print(logit_model.summary())
```

```python
## Problem 3. Examine OLS fitted values
# This section of the code uses the same OLS model to print the predicted values (ols_preds) for each
# data point from the dataset which represent the estimate of loan approval using credit score & DTI 
# predictors
```

```python
ols_preds = ols_model.predict(x)
print(ols_preds)
```

```python
# This section of the code takes our predicted values (ols_preds) and finds how many of those values 
# fall outside of our set # range (0 to 1). Because this is a "yes or no" / binary outcome predictions
# should not fall outside of that range. 
```

```python
out_of_bounds = ols_preds[(ols_preds < 0) | (ols_preds > 1)]
print(out_of_bounds)
print("Number of predictions outside of 0-1 range:", len(out_of_bounds))
```
