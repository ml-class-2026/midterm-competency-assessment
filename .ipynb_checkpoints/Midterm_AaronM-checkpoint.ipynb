```python
import pandas as pd
import statsmodels.api as sm
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
import numpy as np
```

```python
# ── Load the dataset ──────────────────────────────────────────────────────────
```

```python
df = pd.read_csv("loan_approval.csv")

print("Dataset shape:", df.shape)
print("\nFirst five rows:")
print(df.head())
print("\nDescriptive statistics:")
print(df.describe())
```

```python
# ── Problem 1: OLS Regression ─────────────────────────────────────────────────
# Run a linear regression (OLS) with 'approved' as the dependent variable
# and 'credit_score' and 'debt_to_income' as predictors.
# Print the full model summary.
```

```python
X = sm.add_constant(df[['debt_to_income', 'credit_score']])
y = df['approved']

model1 = sm.OLS(y, X).fit()
print(model1.summary())

fitted_vals = model1.fittedvalues
```

```python
# Problem 1: OLS Regression With Test/Train Split -------------------------------
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42)

print("\nTraining set size:", X_train.shape[0], "observations")
print("Testing set size: ", X_test.shape[0], "observations")

X_test_const = sm.add_constant(X_test)
X_train_const  = sm.add_constant(X_train)

model2 = sm.OLS(y_train, X_train_const).fit()
print(model2.summary())

fitted_vals2 = model2.fittedvalues

y_pred = model2.predict(X_test_const)
```

```python
# Problem 1: OLS Regression With Cross Validation -------------------------------
```

```python
ols_cv_scores = cross_val_score(
    LinearRegression(),
    X, y,
    cv=5,
    scoring='r2'
)

print("\n CV R² Scores (each fold):", ols_cv_scores.round(4))
print("Mean CV R²:              ", round(ols_cv_scores.mean(), 4))
print("Std Dev of CV R²:        ", round(ols_cv_scores.std(), 4))
```

```python
# ── Problem 2: Logistic Regression ───────────────────────────────────────────
# Run a logistic regression with the same predictors and outcome variable.
# Print the full model summary.
```

```python
X_const = sm.add_constant(X_train)
logit_sm = sm.Logit(y_train, X_const).fit()

print(logit_sm.summary())
```

```python
# ── Problem 2: Logistic Regression With Cross Validation──────────────────────────────────────────
```

```python
lr_cv_scores = cross_val_score(LogisticRegression(max_iter=1000),
                                X, y, cv=5, scoring='accuracy')
print("CV Scores (each fold):", lr_cv_scores.round(4))
print("Mean CV Accuracy:     ", round(lr_cv_scores.mean(), 4))
print("Std Dev of CV Scores: ", round(lr_cv_scores.std(), 4))
```

```python
# ── Problem 3: OLS Fitted Values ──────────────────────────────────────────────
# Print the fitted values from your OLS model.
# Count and print how many predictions fall outside the range [0, 1].
```

```python
print("Model1 Fitted Values:\n", fitted_vals)
outside_range = ((fitted_vals < 0) | (fitted_vals > 1)).sum()
print("\nNumber of predictions outside [0, 1]:", outside_range)

print("\n Model2 Fitted Values for Train Set :\n", fitted_vals2)
outside_range2 = ((fitted_vals2 < 0) | (fitted_vals2 > 1)).sum()
print("\nNumber of predictions outside [0, 1]:", outside_range2)

print("\n Model 2 Predicted Values For Test Set :\n", y_pred)
outside_range3 = ((y_pred < 0) | (y_pred > 1)).sum()
print("\nNumber of predictions outside [0, 1]:", outside_range3)
```
